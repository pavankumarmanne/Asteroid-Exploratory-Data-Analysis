```python
#eccentricity (e), semi-major axis (a), and inclination (i) of Orbit of a planet
df["Orbit"] = df[["e", "a", "i"]].astype(str).agg(",".join, axis=1)
```

```python
df.dropna(subset = ["prefix"],inplace=True)
```

```python
msno.matrix(df)
plt.figure(figsize=(10,6))
plt.show()
```

```python
df.fillna("Unknown")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['H'].dropna(), kde=True, color='green', bins=30)
plt.title('Distribution of Absolute Magnitude (H)')
plt.xlabel('H (Absolute Magnitude)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for the 'e' column (Eccentricity)
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['e'].dropna(), color='green')
plt.title('Box Plot of Eccentricity (e)')
plt.xlabel('Eccentricity (e)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['a'].dropna(), color='red')
plt.title('Box Plot of Semi-Major Axis(a)')
plt.xlabel('Semi-Major Axis (a)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['i'].dropna(), color='blue')
plt.title('Box Plot of inclanation(i)')
plt.xlabel('inclanation (i)')
plt.show()
```

```python
# Scatter plot for 'a' (Semi-Major Axis) vs 'e' (Eccentricity)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='a', y='e', color='purple')
plt.title('Semi-Major Axis vs Eccentricity')
plt.xlabel('Semi-Major Axis (a)')
plt.ylabel('Eccentricity (e)')
plt.show()
```

```python
# Bar plot to show the distribution of 'class' values
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='class', palette='Set2')
plt.title('Distribution of Celestial Object Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the correlation heatmap for numerical columns
plt.figure(figsize=(12, 8))
sns.heatmap(df[['e', 'a', 'i', 'H', 'n']].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# Plotting the correlation heatmap for numerical columns
plt.figure(figsize=(12, 8))
sns.heatmap(df[['e', 'a', 'i', 'H', 'n']].corr(), annot=True, cmap='crest', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['e'], df['a'], df['i'], c='blue', marker='o')
ax.set_xlabel('Eccentricity (e)')
ax.set_ylabel('Semi-Major Axis (a)')
ax.set_zlabel('Inclination (i)')
ax.set_title('3D Scatter Plot of Orbital Parameters')
plt.show()
```

```python
#The visualizations reveal that most celestial objects in the dataset exhibit a specific range of orbital parameters, with eccentricity showing some outliers. 
#The semi-major axis and eccentricity appear correlated, suggesting typical orbital behavior. 
#The distribution of object classes shows an imbalance, indicating dominance of certain types. 
#Missing values, especially in diameter, need careful handling.
#Overall, the data provides insights into the orbital characteristics, but further imputation and analysis are needed to refine the understanding of celestial objects' properties.
```
